```python
########################
# auteur: kanga elie
########################

###################################
# constante
HEIGHT= 500
WIDTH= 500

###################################
# variables
interupteur = 0
###################################
# importation des bibliotheque
import tkinter as tk


###################################
# fonctions
def color_rect(event):
    global interupteur
    if event.x >= 200 and event.y >= 200 and event.x <= 400 and event.y <= 400 and interupteur == 0 :
        interupteur += 1
        canvas.itemconfig(rectangle, fill= "blue")
    
    elif event.x >= 200 and event.y >= 200 and event.x <= 400 and event.y <= 400 and interupteur == 1 :
        interupteur -= 1
        canvas.itemconfig(rectangle, fill= "red")
    else:
        interupteur = 2

def recommencer():
    global interupteur
    canvas.itemconfig(rectangle, fill= "red")
    interupteur= 0


###################################
# widget principale
racine= tk.Tk()
racine.title("exo 1")
###################################
# les widgets
canvas= tk.Canvas(racine,height=HEIGHT,width=WIDTH, bg="black")
again= tk.Button(racine, text="recommencer", command= recommencer)
rectangle= canvas.create_rectangle((200,200),(400,400),fill="red")
canvas.bind("<Button-1>", color_rect)

###################################
# placement des widgets
canvas.grid(column=0,row=0)
again.grid(column=0,row=1)

###################################
# boucle de l'interface
racine.mainloop()
```
